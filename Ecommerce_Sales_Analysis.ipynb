**Basic Grouping and Single Aggregations**

```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")

```

```python
total_sales=df.groupby('region')['sales_amount'].sum().reset_index()
print(total_sales)
```

```python
no_of_trans=df.groupby('category')['transaction_id'].count().reset_index()
print(no_of_trans)
```

```python
avg_sales=df.groupby('salesperson')['sales_amount'].sum().reset_index()
print(avg_sales)
```

```python
top_sales_region=df.groupby('region')['sales_amount'].sum().sort_values(ascending=False).reset_index()
print(top_sales_region)
```

**Multi-Column Grouping and Multiple Aggregations**

```python
totalsalesmulti=df.groupby(['region','category'])['sales_amount'].sum()
print(totalsalesmulti)
```

```python
salesperson_metrics=df.groupby('salesperson')['sales_amount'].agg(['sum','count','mean']).reset_index()
print(salesperson_metrics)
```

```python
sales_sort=df.groupby('salesperson')['sales_amount'].sum().sort_values(ascending=False).reset_index()
print(sales_sort)
```

```python
max_revenue=df.groupby('category')['sales_amount'].sum().idxmax()
print(max_revenue)
```

**Custom Aggregation and Complete Sales Report**

```python
def sales_range(x):
  return x.max()-x.min()
result=df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print(result)
```

```python
final_summary=df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()
print(final_summary)
```

When we use dictionary based aggregation it take it as different levels that is multiIndex column like already available columns name in level 1 and aggregated columns are in level 2 and if we perform single aggregation it give it as a single level of column.
